## Average Relative price for all Services(Inaptient + Outpatient) between years 2016-2018

```python

# Concatenate all the DataFrames in the list 'df' into a single DataFrame
combined_df = pd.concat(df)

# Calculate the average of the relative prices for each year
avg_prices_2016 = round(combined_df['Relative price for inpatient and outpatient services,2016'].mean(), 2)
avg_prices_2017 = round(combined_df['Relative price for inpatient and outpatient services,2017'].mean(), 2)
avg_prices_2018 = round(combined_df['Relative price for inpatient and outpatient services,2018'].mean(), 2)

# Plot the averages using a bar chart
years = ['2016', '2017', '2018']
average_prices = [avg_prices_2016, avg_prices_2017, avg_prices_2018]

def addlabels(years,average_prices):
    for i in range(len(years)):
        plt.text(i,average_prices[i],average_prices[i])

plt.figure(figsize=(8, 4))
plt.bar(years, average_prices, color=['skyblue'])
plt.xlabel('Year')
plt.ylabel('Average Relative Price')
plt.title('Average Relative Prices for Inpatient and Outpatient Services (2016-2018)')
addlabels(years, average_prices)
# Set y-axis intervals
plt.ylim(230, 270)
plt.show()

```
